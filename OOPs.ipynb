```python
class dog:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def bark(self):
    return f"My {self.age} year old Dog, {self.name} is barking"

my_dog = dog("Harsh", 3)
print(my_dog.bark())
```

```python
class car:
    def __init__(self,brand,modal):
        self.brand = brand
        self.modal = modal
    def carInfo(self):
        print (f"Car Brand is, {self.brand}")
        print (f"Car Modal is, {self.modal}")

my_car = car("Nissan", "R34 GTR")
my_car.carInfo()
```

```python
#Single Inheritance
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, {self.name}")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def greet(self):
        print(f"My Child, {self.name}! You are {self.age} years old.")

my_child = Child("Harsh", 20)
my_child.greet()
```

```python
#Multiple Inheritance
class Parent1:
    def __init__(self, name):
        self.name = name

class Parent2:
    def __init__(self, age):
        self.age = age

class Child(Parent1, Parent2):
    def __init__(self, name, age):
        Parent1.__init__(self, name)
        Parent2.__init__(self, age)

    def greet(self):
        print(f"My Child, {self.name}! You are {self.age} years old.")

my_child = Child("Harsh", 20)
my_child.greet()
```

```python
#Method Overriding
class animal():
  def __init__(self,name):
    self.name = name
  def speak(self):
    print(f"{self.name} makes a sound.")
class dog(animal):
  def _init_(self,name):
    super()._init_(name)
  def speak(self):
    print(f"{self.name} barks.")
a = animal("Genetic Animal")
a.speak()
d = dog("Buddy")
d.speak()
```
