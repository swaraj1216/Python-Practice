```python
#numbers from 1 to 100
number = [num for num in range(1, 101)]
print(number)
```

```python
#even numbers from 1 to 100
number1 = [num for num in range(1, 101) if num%2 == 0]
print(number1)
```

```python
#numbers from 1 to 100 divisible by 7
number2 = [num for num in range(1, 101) if num%7 == 0]
print(number2)
```

```python
#squares of even numbers from 1 to 10
number3 = [num**2 for num in range(1,11) if num%2 == 0]
print(number3)
```

```python
#numbers from 100 to 1
number4 = [num for num in range(100, 0, -1)]
print(number4)
```

```python
#get vowels from the sentence
sentence = "The rocket came back from Mars."
vowels = [char for char in sentence if char.lower() in "aeiou"]
print(vowels)
```

```python
#get consonants from the sentence
sentence = "The rocket came back from Mars."
consonants = [char for char in sentence if char.lower() not in "a eiou."]
print(consonants)
```

```python
#get vowels without repitition from the sentence(set comprehension)
sentence = "The rocket came back from Mars."
vowels = {char for char in sentence if char.lower() in "aeiou"}
print(vowels)
```

```python
original_prices = [1.25, -9.45, 10.22, 3.78, -5.92, 1.16]
formatted_prices = [x if x>0 else 0 for x in original_prices]
print(formatted_prices)
```

```python
#dictionary comprehension, squares from 1 to 5 in key value pair
squares = {key:key**2 for key in range(1,6)}
print(squares)
```

```python
#creating dictionary using key and value lists by dictionary comprehension
keys = ['a', 'b', 'c', 'd']
values = [1, 2, 3, 4]
dict = {keys[i]:values[i] for i in range(len(keys))}
print(dict)
```

```python
#dictionary of square roots of numbers from 100 to 1
roots = {x:round(x**(0.5),2) for x in range(100, 94, -1)}
print(roots)
```

```python
developers = {'Jane': 'Python', 'Jade': 'JavaScript', 'John': 'Python', 'Doe': 'Ruby'}
python_developers = {x:developers[x] for x in developers if developers[x] == 'Python'}
print(python_developers)
```

```python
admins = {'Alice': 'Ubuntu', 'Bob': 'Windows', 'Charlie': 'Ubuntu', 'David': 'MacOS', 'Eve': 'Ubuntu'}
ubuntu_admins = {x:admins[x] for x in admins if admins[x] == 'Ubuntu'}
print(ubuntu_admins)
```
